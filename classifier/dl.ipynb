```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.python.ops.numpy_ops import np_config
np_config.enable_numpy_behavior()
```

```python
data_path = "../data/numbers/"
train, test = keras.utils.image_dataset_from_directory(data_path,labels='inferred', 
                                                label_mode='int',
                                                color_mode='grayscale',
                                                batch_size=32,
                                                seed=0,
                                                image_size=(28,28),
                                                shuffle=True,
                                                subset="both",
                                                validation_split=0.2,
                                                )
```

```python
layers = [keras.layers.Conv2D(filters=3, kernel_size=(3, 3), strides=(1, 1), activation='ReLU'), keras.layers.Conv2D(filters = 1, kernel_size=(3, 3), strides=(1, 1), activation='ReLU'), keras.layers.Flatten(), keras.layers.Dense(10, input_shape=(28*28,), activation='softmax')]
```

```python
layers = [keras.layers.Conv2D(filters=3, kernel_size=(3, 3), strides=(1, 1), activation='ReLU'), keras.layers.Conv2D(filters = 1, kernel_size=(3, 3), strides=(1, 1), activation='ReLU'), keras.layers.Flatten(), keras.layers.Dense(10, input_shape=(28*28,), activation='softmax')]
model = keras.Sequential(layers)
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
def train_model(model, train, test):
    for batch in train:
        x = batch[0]/255
        y = batch[1]
        # x = x.reshape(-1, 28*28)
        model.fit(x, y, validation_split=0.1, epochs=10, steps_per_epoch = 2)
    acc = 0
    n = 0
    for batch in test:
        x = batch[0]/255
        y = batch[1]
        # x = x.reshape(-1, 28*28)
        history = model.evaluate(x, y)
        acc+=history[1]*len(batch)
        n+=len(batch)
    accuracy = acc/n  
    print("Test accuracy : ", accuracy)  
        
```

```python
train_model(model, train, test)
```

```python
model.save(filepath="../models/num_pred.h5", overwrite=True)
```
